# Topic Modeling on ArXiv Abstracts

Before topic modeling we will need access to data. Below I've used the arXiv api to grab the first 20 results found for the `cs.AI` category.

```python
import urllib, urllib.request

url = 'http://export.arxiv.org/api/query?search_query=cat:cs.AI&start=0&max_results=20'
data = urllib.request.urlopen(url)
xml_content = data.read().decode('utf-8')
print(xml_content)
```

The API response comes in XML format and the abstract for each entry is surrounded by tags: `<summary>Abstract</summary>`.
We can extract the abstracts in the XML response but we'll need the `namespace` which you can think of as an identifier for the specific XML format/schema being used. It's specified in the XML output near the top - `<feed xmlns="http://www.w3.org/2005/Atom">`.

```python
import xml.etree.ElementTree as ET
# Parse the XML
root = ET.fromstring(xml_content)

# Define namespace
namespace = {
    'atom': 'http://www.w3.org/2005/Atom'
}

# Get ALL entries (not just the first one)
entries = root.findall('atom:entry', namespace)

# Loop through each entry and extract the abstract
for i, entry in enumerate(entries, 1):
    abstract = entry.find('atom:summary', namespace).text
    title = entry.find('atom:title', namespace).text
    print(f"\n--- Paper {i} ---")
    print(f"Title: {title}")
    print(f"Abstract: {abstract[:200]}...")
```
